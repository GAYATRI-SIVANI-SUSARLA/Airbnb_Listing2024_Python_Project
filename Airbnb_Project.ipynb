```python
##4. Data Cleaning 
```

```python
## check for sum of null values in every column 
data.isnull().sum()

##dropping all missing values
data.dropna(inplace = True)

data.isnull().sum()
```

```python
##check data has duplicate values 
data.duplicated().sum() 

##list duplicated records
data[data.duplicated()]

## deleting all duplicated rows or records
data.drop_duplicates(inplace=True) 
data.duplicated().sum()
```

```python
##type casting
##data types of columns
data.dtypes

##change 'id' to object
data['id'] = data['id'].astype(object)
data.dtypes

##change 'host_id' to object
data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

```python
##EDA
## 5. Data Analysis 
##Univariate Analysis- Analyzing every single column and its distribution
```

```python
##Create dataframe to remove outliers 
df = data[data['price'] < 1500]


##Identifying outliers in the price 
sns.boxplot(data = df, x ='price')

```

```python
##Price distribution
plt.figure(figsize = (8,5))
sns.histplot(data=df, x= 'price',bins=50)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
## list out all columns
df.columns
```

```python
##Availability_365 distribution
plt.figure(figsize = (6,3))
sns.histplot(data=df, x= 'availability_365',bins=20)
plt.title("Availability_365 Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
##Analyse neighbourhood_group by average price
df.groupby(by ='neighbourhood_group')['price'].mean()
```

```python
##Feature Engineering
##price per bed 
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
##Price per bed group by neighbourhood group
df.groupby('neighbourhood_group')['price per bed'].mean()
```

```python
##Bivariate Analaysis
##One variable dependency on other variables 
```

```python
##price dependency on neighbourhood_group
sns.barplot(data = df, x='neighbourhood_group', y='price',hue='room_type')
```

```python
 ##Number of reviews given with respective price 
plt.figure(figsize=(10,5))
plt.title("Relation between Reviews & Price given by Customers")
sns.scatterplot(data= df,x ='number_of_reviews',y = 'price', hue= 'neighbourhood_group')
plt.show()
```

```python
##dependencies with other columns
df.dtypes
sns.pairplot(data = df, vars = ['price', 'minimum_nights', 'number_of_reviews','availability_365'],hue = 'room_type')
```

```python
## Geographical distribution of Airbnb Listing
plt.figure(figsize= (10,8))
plt.title("Geographical distribution of Airbnb")
sns.scatterplot(data = df, x = 'latitude', y = 'longitude',hue='room_type')
plt.show()
```

```python
## Heat Map = correlation of one variable with other numerical columns 
corr = df[['latitude', 'longitude', 'price', 'minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()                             
corr
plt.title("Correlation between Numeric variables")
sns.heatmap(data = corr, annot = True)
plt.show()
```
